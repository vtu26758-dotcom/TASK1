```python
import pandas as pd
import numpy as np
from scipy import stats
import seaborn as sns

df = sns.load_dataset("tips")

print("Dataset Info:")
df.info()
print("\nFirst 5 rows:\n", df.head())

print("\nSummary Statistics:\n", df.describe())

print("\nCentral Tendency (total_bill) using NumPy:")
print("Mean:", np.mean(df['total_bill']))
print("Median:", np.median(df['total_bill']))
print("Mode:", stats.mode(df['total_bill'], keepdims=False).mode)

print("\nCentral Tendency (total_bill) using Pandas:")
print("Mean:", df['total_bill'].mean())
print("Median:", df['total_bill'].median())
print("Mode:", df['total_bill'].mode().values[0])

print("\nGender Distribution:\n", df['sex'].value_counts())
print("\nDay Distribution:\n", df['day'].value_counts())

print("\nAverage Tip by Gender:\n", df.groupby('sex')['tip'].mean())
print("\nAverage Total Bill by Day:\n", df.groupby('day')['total_bill'].mean())

print("\nCorrelation Matrix:\n", df.corr(numeric_only=True))
```
