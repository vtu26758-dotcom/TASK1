```python
from PIL import Image
import pandas as pd
import requests
import csv

def process_image(image_path):
    img = Image.open("C:\\Users\\M siddharth\\OneDrive\\Pictures\\siddu.jpg")
    print("Image Details:")
    print("Format:", img.format)
    print("Size:", img.size)
    print("Mode:", img.mode)

def process_csv(csv_path):
    df = pd.read_csv("C:\\Users\\M siddharth\\OneDrive\\Desktop\\students.csv")
    print("\n CSV File Columns:", df.columns.tolist())
    print("Data Types:\n", df.dtypes)
    return df

def process_excel(excel_path):
    df = pd.read_excel("C:\\Users\\M siddharth\\OneDrive\\Desktop\\student_marks.xlsx")
    print("\n Excel File Preview:\n", df.head())
    df['Average'] = df.iloc[:, 1:].mean(axis=1)
    return df

def process_json_from_api(url):
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        print("\n JSON from API (https://disease.sh):")
        print("Updated:", data['updated'])
        print("Total Cases:", data['cases'])
        print("Deaths:", data['deaths'])
        print("Recovered:", data['recovered'])
        return {
            "TotalCases": data['cases'],
            "Deaths": data['deaths'],
            "Recovered": data['recovered']
        }
    else:
        print("Failed to fetch API data.")
        return {}

def save_summary_to_csv(df, summary_data, output_path):
    df.to_csv(output_path, index=False)
    print(f"\n Processed Excel data saved to: {output_path}")
    with open(output_path, 'a', newline='') as file:
        writer = csv.writer(file)
        writer.writerow([])
        writer.writerow(['--- COVID-19 Global Summary ---'])
        for key, value in summary_data.items():
            writer.writerow([key, value])
    print(" COVID-19 summary appended to the same CSV.")

image_path = "C:\\Users\\M siddharth\\OneDrive\\Pictures\\siddu.jpg"
csv_path ="C:\\Users\\M siddharth\\OneDrive\\Desktop\\students.csv"
excel_path = "C:\\Users\\M siddharth\\OneDrive\\Desktop\\student_marks.xlsx"
json_url = 'https://disease.sh/v3/covid-19/all'
output_csv ="C:\\Users\\M siddharth\\Downloads\\sid.csv"
process_image("C:\\Users\\M siddharth\\OneDrive\\Pictures\\siddu.jpg")
csv_data = process_csv("C:\\Users\\M siddharth\\OneDrive\\Desktop\\students.csv")
excel_data = process_excel("C:\\Users\\M siddharth\\OneDrive\\Desktop\\student_marks.xlsx")
json_summary = process_json_from_api('https://disease.sh/v3/covid-19/all')
save_summary_to_csv(excel_data, json_summary, output_csv)
```
